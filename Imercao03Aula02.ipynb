### Desafio 01: Ordernar o gráfico countplot

```python
sns.set()
plt.figure(figsize=(8,6))
#plt.title('')
ax = sns.countplot(x= 'composto', data=dados.query('composto in @cod_compostos'),order=cod_compostos)
ax.set_title('Top 5 Compostos')
```

### Desafio 02: Melhorar a visualização alterando tamanho da fonte.

```python

plt.figure(figsize=(8,6))
plt.title('Top 5 Compostos',fontsize=16, fontweight = "bold")
ax = sns.countplot(x= 'composto', 
                   data=dados.query('composto in @cod_compostos'),
                   order=cod_compostos,
                   edgecolor=sns.color_palette("dark", 5))

plt.ylabel('Valor', fontsize=12, fontweight = "bold")
plt.xlabel('Composto', fontsize=12, fontweight = "bold")

plt.show()
```
